### preparing the data

```python
lc=text.lower()
```

```python
lc
```

#### Removing punctuation

```python
import string
ct=lc.translate(str.maketrans('','',string.punctuation))
```

```python
ct
```

```python
nltk.download('punkt')
```

#### Tokenizing the data

```python
token_words=word_tokenize(ct,"english")
```

```python
token_words
```

```python
from nltk.corpus import stopwords
```

#### Removing stopwords

```python
nltk.download('stopwords')
```

```python
final=[]
for word in token_words:
    if word not in stopwords.words('english'):
        final.append(word)
```

```python
final
```

```python
from collections import Counter
```

#### Collecting emotion

```python
emolist=[]
with open("emotions.txt","r") as file2:
    for line2 in file2:
        clear_line2= line2.replace('\n','').replace(",","").replace("'","").strip()
        #print(clear_line2)
        word, emotion= clear_line2.split(":")
        #print("Word:"+word+" "+"Emotions:"+emotion)
        if word in final:
            emolist.append(emotion)
```

```python
emolist
```

#### Counting different emotions

```python
w=Counter(emolist)
```

```python
w
```

### Visualization of data

```python
import matplotlib.pyplot as plt
plt.bar(w.keys(),w.values())
plt.savefig("emo.png")
plt.show()
```

```python
#for automatic tiltation
fig, axl=plt.subplots()
axl.bar(w.keys(),w.values())
fig.autofmt_xdate() #update x axis into tilted format
plt.show()
```

#### Showcasing the results

```python
def sentiment_analise(sentext):
    score=SentimentIntensityAnalyzer().polarity_scores(sentext)
    neg=score["neg"]
    pos=score["pos"]
    if neg > pos:
        print("negetive")
    elif pos > neg:
        print("positive")
    else:
        print("neutral bitch")
```
